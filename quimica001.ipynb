```python
import pandas as pd
import numpy as np

# ===== arquivo =====
arquivo = "/content/Tabela analise multivariada_FJMN (2).xlsx"

# =========================================================
# 1) TABELA QUÍMICA: componentes dos óleos
# =========================================================
x_raw = pd.read_excel(arquivo, sheet_name="Componentes dos óleos", header=None)

# nomes das amostras estão na linha 1, mas há colunas vazias entre elas
linha_amostras = x_raw.iloc[1, 2:].tolist()
amostras = [v for v in linha_amostras if pd.notna(v)]
cols_validas = [i for i, v in enumerate(linha_amostras, start=2) if pd.notna(v)]

# monta tabela: 1 coluna de composto + colunas das amostras
x = x_raw.iloc[2:, [1] + cols_validas].copy()
x.columns = ["Composto"] + amostras

# converte para formato amostra x composto
x = x.set_index("Composto")
x = x.apply(pd.to_numeric, errors="coerce")
x = x.T.reset_index().rename(columns={"index": "Código"})

print("Tabela química:")
print(x.shape)
display(x.head())

# =========================================================
# 2) TABELA BIOLÓGICA: cicatrização (área da ferida) - médias
# =========================================================
y_raw = pd.read_excel(arquivo, sheet_name="Cicatrização (área da ferida)", header=None)

# tabela de médias está na parte esquerda
y = y_raw.iloc[3:, 0:4].copy()
y.columns = ["Código", "Dia_zero", "Dia_4", "Dia_8"]

# remove linhas vazias
y = y.dropna(subset=["Código"])

# converte valores para numérico
for c in ["Dia_zero", "Dia_4", "Dia_8"]:
    y[c] = pd.to_numeric(y[c], errors="coerce")

print("Tabela biológica:")
print(y.shape)
display(y.head())

# =========================================================
# 3) JUNÇÃO QUÍMICA + BIOLÓGICA
# =========================================================
df = pd.merge(x, y, on="Código", how="inner")

# cria variável de contração no dia 8
df["Contracao_D8"] = (df["Dia_zero"] - df["Dia_8"]) / df["Dia_zero"]

print("Tabela final unificada:")
print(df.shape)
display(df.head())

print("Colunas:")
print(df.columns.tolist())
```

```python
from sklearn.cross_decomposition import PLSRegression
from sklearn.model_selection import LeaveOneOut
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.preprocessing import StandardScaler
import numpy as np
import pandas as pd

# ==============================
# 1) definir X e y
# ==============================

y = df["Contracao_D8"]

cols_meta = ["Código","Dia_zero","Dia_4","Dia_8","Contracao_D8"]
X = df.drop(columns=cols_meta)

# ==============================
# 2) padronizar
# ==============================

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# ==============================
# 3) modelo PLS
# ==============================

pls = PLSRegression(n_components=3)

loo = LeaveOneOut()

y_real = []
y_pred = []

for train, test in loo.split(X_scaled):

    X_train, X_test = X_scaled[train], X_scaled[test]
    y_train, y_test = y.iloc[train], y.iloc[test]

    pls.fit(X_train, y_train)

    pred = pls.predict(X_test)

    y_real.append(float(y_test))
    y_pred.append(float(pred))

# ==============================
# 4) métricas
# ==============================

r2 = r2_score(y_real, y_pred)
rmse = np.sqrt(mean_squared_error(y_real, y_pred))

print("R2:", r2)
print("RMSE:", rmse)

resultado = pd.DataFrame({
    "real": y_real,
    "predito": y_pred
})

display(resultado)
```
